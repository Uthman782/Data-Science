```python
df = pd.read_csv('E://weatherData.csv')
df
```

```python
g = df.groupby('city')
g
```

```python
swat = g.get_group('swat')
swat
```

```python
mardan = g.get_group('mardan')
mardan
```

```python
kanju =g.get_group('kanju')
kanju
```

```python
# df_2 = pd.concat([swat, mardan], ignore_index=True)
df_2 = pd.concat([swat, mardan], keys = ['Swat', 'Mardan'])
df_2
```

```python
df_3 = pd.concat([swat, kanju], ignore_index=True)
# df_3.max()
df_3
```

```python
df_4 = pd.DataFrame({
    'city':['swat', 'kanju', 'mardan'],
    'temperature':[24,26,28],
}, index=[0,1,2])
df_4
```

```python
df_5 = pd.DataFrame({
    'city':['kajus', 'mardan', 'swat'],
    'windspeed':[38,25,55]
}, index=[1,2,0])
df_5
```
