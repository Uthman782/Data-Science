```python
import pandas as pd
```

```python
df_1 = pd.DataFrame({
    'city':['swat', 'kanju', 'mardan', 'lahour'],
    'temperature':[24,26,28,38],
})
df_1
```

```python
df_2 = pd.DataFrame({
    'city':['kanju', 'mardan', 'swat', 'karachi'],
    'windspeed':[38,25,55, 45]
})
df_2
```

```python
df_3 = pd.merge(df_1,df_2, on='city', how='outer', indicator=True)
df_3
```

```python
df_3 = pd.merge(df_1,df_2, on = 'city', how = 'inner', indicator=True)
df_3
```

```python
df_3 = pd.merge(df_1, df_2, on= 'city', how = 'right', indicator=True)
df_3
```

```python
df_3 = pd.merge(df_1, df_2, on= 'city', how = 'left', indicator=True)
df_3
```
