## Pyplot vs Matplotlib OO method

```python
over_50=df[df['age']>50]
over_50.head()
```

```python
# pyplot method
ax=over_50.plot(kind='scatter', x='age', y='chol', c='target', figsize=(12, 6), cmap='plasma');
ax.set_xlim([50, 80])
# fig, ax=plt.subplots(figsize=(12, 6))
# scatter = ax.scatter(x=over_50['age'],
#                      y=over_50['chol'],
#                      c=over_50['target'],
#                      cmap='winter')
# ax.set_xlim([50, 80])
```

```python
# OO method mixed with pyplot
fig, ax=plt.subplots(figsize=(12, 6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax = ax, cmap='summer');
ax.set_xlim([45, 100]);
```

```python
# OO method from scratch
fig, ax =plt.subplots(figsize=(10, 6))
                      
scatt = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
ax.set(title='Heart Disease and cholesterol Levels',
       xlabel='Age',
       ylabel='chol')
ax.legend(*scatt.legend_elements(), title='Target',fontsize='larger')
ax.axhline(over_50['chol'].mean(), linestyle='--');
```

```python
fig, (ax0, ax1)=plt.subplots(nrows=2,
                             ncols=1,
                            figsize=(10,12))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax0)
over_50.plot(kind='scatter', x='age',y='thalach', c='target' , ax=ax1);
ax0.set(title='Chol Level')
ax1.set(title='Thalach Level');
```

```python
fig, (ax0, ax1)=plt.subplots(figsize=(10,12),
                             nrows=2,
                             ncols=1,
                             sharex=True)
# Add Data to ax0
scatter=ax0.scatter( x=over_50['age'], # We shall Share X (axis)
                      y=over_50['chol'],
                      c=over_50['target'])
# Customizing ax0
ax0.set(title='Heart Disease and Cholestrol Level',
        # xlabel='Age', # It will Sharing X label
        ylabel='Cholesterol');
ax0.axhline(over_50['chol'].mean(), linestyle='--')
ax0.legend(*scatter.legend_elements(), title='Target');

# Add Data to ax1
scatter=ax1.scatter(x=over_50['age'], 
                      y=over_50['thalach'],
                      c=over_50['target'])
# Customizing ax1
    
ax1.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='Thalach Rate');
ax1.axhline(over_50['thalach'].mean(), linestyle='--')
ax1.legend(*scatter.legend_elements(), title='Target')
fig.suptitle('Heart Disease Analysis', fontsize=15, fontweight='bold');
```
