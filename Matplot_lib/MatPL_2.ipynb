```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=range(1,8)
y=[20,22,24,25,21,20,19]
```

```python
plt.plot(x,y, 'D--', color='#eeaa0e')
```

```python
plt.plot(x,y, 'r*-')
plt.show()
```

```python
plt.plot(x,y, 'gD-')
```

```python
plt.plot(x,y, color='red', marker='*', linestyle='dotted', markersize=15)
```

```python
plt.plot(x,y, color='red', marker='*', linestyle='dotted', markersize=15, alpha=0.4)
```

```python
plt.plot(x,y, color='#00aaee', marker='+', linestyle='dotted', markersize=15, animated=True)
```

```python
plt.plot(x,y, color='black', marker='x', linestyle='dotted', markersize=15, )
```
