```python
df=pd.read_csv('E:\\Local Disk E\\Data Scientist\\Complete Machine Learning & Data Science Bootcamp 2021\\8. Matplotlib Plotting and Data Visualization\\13.1 heart-disease.csv')
over_50=df[df['age']>50]
over_50.head()
```

```python
plt.style.use('fivethirtyeight')
fig, ax=plt.subplots(figsize=(10,6))
scatter=ax.scatter( x=over_50['age'], # We shall Share X (axis)
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter')
# Customizing ax
ax.set(title='Heart Disease and Cholesterol Level',
        # xlabel='Age', # It will Sharing X label
        ylabel='Cholesterol');
ax.axhline(over_50['chol'].mean(), linestyle='--')
ax.legend(*scatter.legend_elements(), title='Target');
```
