```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4]
y = [11,22,33,44]
z = [44,33,22,11]

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, y, label='first', color='#33aaff')
ax.plot(x, z, label='second', color='#00ffaa')
ax.set(title='Simple Plot',
       xlabel='x-axis',
       ylabel='y-axis')
ax.grid()
ax.legend(fontsize='large')
plt.show()
```

```python
x = range(1,8)
y = [15,17,18,22,19,16,20]
```

```python
fig, ax = plt.subplots(figsize=(13,5))
ax.plot(x,y, color='red')
ax.set(title='Weather', xlabel='days', ylabel='Temperature')
plt.grid()
plt.show()
```

```python
import numpy as np
x = np.linspace(0,10,100)
x[0:10]
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(x, x**2)
ax.grid()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(x, np.sin(x), color='#20ffee', label='SIN')
ax.scatter(x, np.cos(x), color='#1eaaff', label='COS')
ax.grid()
ax.legend();
```

```python
fig, ax=plt.subplots(figsize=(10,5))
ax.bar(x,np.sin(x), color="red")
ax.bar(x,np.cos(x), color='#33eeaa')
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x), color='#aaccee');
```

```python
deshes_prices = {'Bread':50,
                 'Beef': 150,
                 'Curd': 120,
                 'Milk': 140,
                 'Whey': 110}
```
