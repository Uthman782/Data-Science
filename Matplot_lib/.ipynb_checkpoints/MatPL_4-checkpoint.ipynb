```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
company = ['Google', 'Amzon','Msft', 'FB']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos, company)
plt.ylabel('revenue(bln)')
plt.title('Us Tech Stocks')
plt.bar(xpos, revenue, label = 'Rvenue')
plt.bar(xpos, profit, label='Profit')
plt.legend()
```

```python
plt.xticks(xpos, company)
plt.ylabel('revenue(bln)')
plt.title('Us Tech Stocks')
plt.bar(xpos-0.2, revenue, label = 'Rvenue')
plt.bar(xpos+0.2, profit, label='Profit')
plt.legend()
plt.show()
```

```python
plt.xticks(xpos, company)
plt.ylabel('revenue(bln)')
plt.title('Us Tech Stocks')
plt.bar(xpos-0.2, revenue, width=0.3, label = 'Rvenue')
plt.bar(xpos+0.2, profit, width=0.3, label='Profit')
plt.legend()
```

```python
plt.yticks(xpos, company)
plt.xlabel('revenue(bln)')
plt.title('Us Tech Stocks')
plt.barh(xpos+0.2, revenue,  height=0.4, label = 'Rvenue', color='#aa02ea')
plt.barh(xpos-0.2, profit,  height=0.4, label='Profit', color = 'orange')
plt.legend()
```
