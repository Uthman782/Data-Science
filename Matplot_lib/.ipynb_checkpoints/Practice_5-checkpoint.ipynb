```python
df=pd.read_csv('E:\\Local Disk E\\Data Scientist\\Complete Machine Learning & Data Science Bootcamp 2021\\8. Matplotlib Plotting and Data Visualization\\13.1 heart-disease.csv')
over_50=df[df['age']>50]
over_50.head()
```

```python
plt.style.use('seaborn-dark')
fig, (ax0, ax1)=plt.subplots(figsize=(15,12),
                             nrows=2,
                             ncols=1,
                             sharex=True)
# Add Data to ax0
scatter=ax0.scatter( x=over_50['age'], # We shall Share X (axis)
                      y=over_50['chol'],
                      c=over_50['target'],
                      cmap='winter')
# Customizing ax0
ax0.set(title='Heart Disease and Cholestrol Level',
        # xlabel='Age', # It will Sharing X label
        ylabel='Cholesterol');
ax0.axhline(over_50['chol'].mean(), linestyle='--')
ax0.legend(*scatter.legend_elements(), title='Target');

# Add Data to ax1
scatter=ax1.scatter(x=over_50['age'], 
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap='plasma')
# Customizing ax1
    
ax1.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='Thalach Rate');
ax1.axhline(over_50['thalach'].mean(), linestyle='--')
ax1.legend(*scatter.legend_elements(), title='Target')
fig.suptitle('Heart Disease Analysis', fontsize=15, fontweight='bold');
```
