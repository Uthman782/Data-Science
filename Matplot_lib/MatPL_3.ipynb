```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1,2,3,4,5,6,7]
max_t = [50,51,52,48,47,49,46]
min_t = [43,42,40,44,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,max_t, label='Max')
plt.plot(days,min_t, label='Min')
plt.plot(days,avg_t, label='Avg')
# plt.legend(loc = 'upper left')
plt.legend(loc='best', shadow = True, fontsize = 'large')
# plt.show()
plt.savefig('C:\\Users\\Usman Khan\\Downloads\\myFig.jpg', bbox_inches='tight', pad_inches=0, transparent = True)
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(days,avg_t, color = 'red', label='Avg')
plt.plot(days,max_t, color = '#00ffaa', label='Max')
plt.plot(days,min_t, color = '#ff00aa', label='Min')
plt.legend(loc = 'best', fontsize = 'large', shadow = True)
plt.grid()
# plt.savefig('C:\\User\\Usman Khan\\Downlpads\\myFig2.png', bbox_inches = 'tight')
```
