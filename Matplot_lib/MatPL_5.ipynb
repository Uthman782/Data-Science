```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113,85,90,150,149,88,93,115,135,80,77,82,129]
# plt.hist(blood_sugar, bins = [80,100,125,150], rwidth = 0.60, histtype='step', color = 'g')
plt.hist(blood_sugar, bins = [80,100,125,150], rwidth = 0.60, color = 'g')
```

```python
blood_sugar_men = [113,85,90,150,149,88,93,115,135,80,77,82,129]
blood_sugar_women = [67,98,89,120,133,150,84,69,89,79,120,112,100]
plt.xlabel('Suger Range')
plt.ylabel('Total No of Patients')
plt.title('Blood Suger Analysis')
plt.hist([blood_sugar_men, blood_sugar_women], bins = [80,100,125,150], rwidth = 0.60, color = ['#00bbee','#11ffee'],
        label=['Men', 'Women'])
plt.legend()
plt.show()
```
