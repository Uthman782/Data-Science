```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4]
y = [11,22,33,44]
z = [44,33,22,11]

fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, y, label='first', color='#33aaff')
ax.plot(x, z, label='second', color='#00ffaa')
ax.set(title='Simple Plot',
       xlabel='x-axis',
       ylabel='y-axis')
ax.grid()
ax.legend(fontsize='large')
plt.show()
```

## Histogram

```python
x = np.random.randn(1000)
fig, ax = plt.subplots(figsize=(10, 6))
ax.hist(x, bins=20);
ax.hist(np.sin(x), bins=20)
ax.hist(np.random.random(20), bins=20);
ax.grid();
```

## Tow options for subplots
### option One

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(15,7))
ax1.grid()
```

```python
x = [1.8,3,5,3,4,6,4,2.5,2.8,4,6,4,6,4,5]
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(15,7))
ax1.plot(x, color='#0accdd')
ax1.bar(np.arange(1,16), x)
ax2.scatter(np.arange(1,16), x)
ax3.bar(deshes_prices.keys(), deshes_prices.values())
ax4.hist(np.random.randn(100));
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(16,8))
ax1.plot(np.arange(1,16),x, color='#00ffff')
ax1.bar(np.arange(1,16), x)
ax1.set(title='This is First Graph')
ax2.plot(np.arange(1,16), x, c='#342311')
ax2.scatter(np.arange(1,16), x)
ax2.set(title='This is Second Graph')
ax3.plot(deshes_prices.values(), c='#998877')
ax3.bar(deshes_prices.keys(), deshes_prices.values())
ax4.hist(np.random.randn(100))
ax3.set(title='This is Third Graph')
ax4.set(title='This is Forth Graph');
ax1.grid()
ax2.grid()
ax3.grid()
ax4.grid()
```

```python
fig, ax=plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(20,10))
ax[0, 0].plot(np.arange(1, 16), x)
ax[0, 0].bar(np.arange(1, 16) , x, color='#aaffff')
ax[0, 0].grid()
ax[0, 1].bar(np.arange(1,8), [5,9,11,19,10,20,9])
ax[0, 1].set(title='Temperature', xlabel='Days', ylabel='Temperature')
ax[0, 1].grid()
ax[0, 1].plot(np.arange(1, 8), [5,9,11,19,10,20,9], color='#ee8765')
ax[1, 0].scatter(np.random.random(100), np.random.random(100))
ax[1, 0].grid()
ax[1, 1].hist(np.random.randn(100))
ax[1, 1].grid()
```
