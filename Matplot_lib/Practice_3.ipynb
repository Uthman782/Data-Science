```python
df_1 = df[:]
df_1['Price']=df_1['Price'].str[:-2]
df_1
```

```python
df_1['Sale Date']=pd.date_range("1/1/2020", periods=(len(df_1['Price'])))
df_1
```

```python
df_1['Total Sale']=df_1['Price'].astype(int).cumsum()
df_1
```

```python
df_1.plot(x='Sale Date', y='Total Sale');
```
