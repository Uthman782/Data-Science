```python
import pandas as pd
```

```python
df = pd.read_csv('E:\\Local Disk E\\Data Scientist\\excel\\home_prices_1.csv')
df
```

```python
dumies = pd.get_dummies(df.town)
dumies
```

```python
dumies_df = pd.concat([df, dumies], axis='columns')
dumies_df
```

```python
dumies_df.drop(['town', 'lahour'], axis='columns', inplace=True)
dumies_df
```

```python
x = dumies_df.drop('price', axis='columns')
x
```

```python
y = dumies_df.price
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x, y)
```
